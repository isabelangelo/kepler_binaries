```python
# I think I was able to get a good plot of the original model
# for order 1.
import pandas as pd
import thecannon as tc
from pylab import *
from astropy.io import fits

# load wavelength
w_order1 = fits.open('./data/cannon_training_data/cannon_reference_w.fits')[0].data[0]

# load spectra from training set
training_flux_original = pd.read_csv('./data/cannon_training_data/training_flux_original.csv')
training_sigma_original = pd.read_csv('./data/cannon_training_data/training_sigma_original.csv')
training_flux_dwt = pd.read_csv('./data/cannon_training_data/training_flux_dwt.csv')
training_sigma_dwt = pd.read_csv('./data/cannon_training_data/training_sigma_dwt.csv')

# only look at order 1
training_flux_original_order1 = training_flux_original.query('order_number==1')
training_sigma_original_order1 = training_sigma_original.query('order_number==1')
training_flux_dwt_order1 = training_flux_dwt.query('order_number==1')
training_sigma_dwt_order1 = training_sigma_dwt.query('order_number==1')

# training set labels
training_labels = pd.read_csv('./data/label_and_metric_dataframes/training_labels.csv')

# trained cannon models
original_model = tc.CannonModel.read('./data/cannon_models/order1_original/cannon_model.model')
dwt_model = tc.CannonModel.read('./data/cannon_models/order1_dwt/cannon_model.model')

# keys to store for comparison
keys = ['smemp_teff', 'smemp_logg','smemp_feh', 'smemp_vsini',\
        'cannon_teff', 'cannon_logg', 'cannon_feh', 'cannon_vsini']

# function to plot order stats
def rms(x, y):
    diff = y - x
    rms = np.round(np.sqrt(np.sum(diff**2)/len(diff)), 3)
    return rms
```

```python
# the code is doing much better here than the one2one says...
# let me generate a one2one plot with the first 20 training set stars.

original_label_dict_list = []
for star in training_labels.id_starname:
    star_flux = training_flux_original_order1[star]
    star_sigma = training_sigma_original_order1[star]

    # true model
    row = training_labels[training_labels.id_starname==star].iloc[0]

    # best-fit model
    fit_result = original_model.test(star_flux, 1/star_sigma**2)[0]
    values = [row.smemp_teff, row.smemp_logg, row.smemp_feh, row.smemp_vsini,\
    fit_result[0][0], fit_result[0][1], fit_result[0][2], fit_result[0][3]]
    original_label_dict_list.append(dict(zip(keys, values)))
    
original_df = pd.DataFrame(original_label_dict_list)

```

```python
figure(figsize=(15,5))

subplot(141)
plot(original_df.smemp_teff, original_df.cannon_teff, 'o')
text(5000,6250, 'rms = {}'.format(rms(original_df.smemp_teff, original_df.cannon_teff)))

subplot(142)
plot(original_df.smemp_logg, original_df.cannon_logg, 'o')
text(3,4.5, 'rms = {}'.format(rms(original_df.smemp_logg, original_df.cannon_logg)))

subplot(143)
plot(original_df.smemp_feh, original_df.cannon_feh, 'o')
text(-0.4,0.4, 'rms = {}'.format(rms(original_df.smemp_feh, original_df.cannon_feh)))

subplot(144)
plot(original_df.smemp_vsini, original_df.cannon_vsini, 'o')
text(2,10, 'rms = {}'.format(rms(original_df.smemp_vsini, original_df.cannon_vsini)))
```

```python
# okay...the test step on the original model training set seems to work well
# now let me run the test step on the dwt model training set.

dwt_label_dict_list = []
for star in training_labels.id_starname:
    star_flux = training_flux_dwt_order1[star]
    star_sigma = training_sigma_dwt_order1[star]

    # true model
    row = training_labels[training_labels.id_starname==star].iloc[0]

    # best-fit model
    fit_result = dwt_model.test(star_flux, 1/star_sigma**2)[0]
    values = [row.smemp_teff, row.smemp_logg, row.smemp_feh, row.smemp_vsini,\
    fit_result[0][0], fit_result[0][1], fit_result[0][2], fit_result[0][3]]
    dwt_label_dict_list.append(dict(zip(keys, values)))
    
dwt_df = pd.DataFrame(dwt_label_dict_list)
```

```python
figure(figsize=(15,5))

subplot(141)
plot(dwt_df.smemp_teff, dwt_df.cannon_teff, 'bo')
title('rms = {}'.format(rms(dwt_df.smemp_teff, dwt_df.cannon_teff)))
plot([4500,7000],[4500,7000],'k-')

subplot(142)
plot(dwt_df.smemp_logg, dwt_df.cannon_logg, 'bo')
title('rms = {}'.format(rms(dwt_df.smemp_logg, dwt_df.cannon_logg)))
plot([2.5,5],[2.5,5],'k-')

subplot(143)
plot(dwt_df.smemp_feh, dwt_df.cannon_feh, 'bo')
title('rms = {}'.format(rms(dwt_df.smemp_feh, dwt_df.cannon_feh)))
plot([-1,0.5],[-1,0.5],'k-')

subplot(144)
plot(dwt_df.smemp_vsini, dwt_df.cannon_vsini, 'bo')
title('rms = {}'.format(rms(dwt_df.smemp_vsini, dwt_df.cannon_vsini)))
plot([0,15],[0,15],'k-')
```
